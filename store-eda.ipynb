```python
df = pd.read_csv("data/store.csv", sep = "|")
print(df.shape)
df.head()
```

```python
pd.options.display.max_columns = 39
df.head()
```

```python
df.info()
```

```python
df_null = df.isnull().sum()
df_null
```

```python
df_null.sort_values().plot.barh(figsize = (7, 8))
```

```python
null_cols = df_null.sort_values(ascending = False).head(9).index
df = df.drop(null_cols, axis = 1)
not_use_cols = df.columns[df.columns.str.contains("코드|번호")]
df = df.drop(not_use_cols, axis = 1)
```

```python
df.shape
```

```python
df.head(3)
```

```python
df.describe()
```

```python
sns.distplot(df["위도"])
plt.axvline(df["위도"].mean(), linestyle = ":", color = "r")
plt.axvline(df["위도"].median(), linestyle = "--")
```

```python
corr = df.corr()
corr
```

```python
import numpy as np
sns.heatmap(corr, annot = True, cmap = "Blues", mask = np.triu(np.ones_like(corr, dtype = np.bool)))
```

```python
sns.lmplot(data = df.sample(100), x = "경도", y = "위도", hue = "시도명", col = "시도명")
```

```python
df.head()
```

```python
df_food = df[(df["시도명"] == "서울특별시") & (df["상권업종대분류명"] == "음식")].groupby(["시군구명", "상권업종중분류명"])["상호명"].count()
df_food = df_food.reset_index().rename(columns = {"상호명" : "상호수"})
df_food
```

```python
df_food[df_food["시군구명"] == "강남구"].plot(kind = "bar", x = "상권업종중분류명")
```

```python
sns.catplot(data = df_food, kind = "bar", x = "시군구명", y = "상호수", col = "상권업종중분류명", col_wrap = 4)
```
